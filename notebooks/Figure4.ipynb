```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from matplotlib.ticker import ScalarFormatter
```

```python
ig_data = pd.read_csv('../summary_data/Figure4a_data.csv')
ig_data
```

```python
colors = ['#F7941D', '#00b9f2', '#00a875']
sns.set_palette(sns.color_palette(colors))
```

```python
plt.figure(figsize=(8,6))

sns.set_style('white')
sns.set_context('talk')

g = sns.lineplot(data=ig_data, x='ind_sharing', y='num_cdr3', hue='IG', hue_order=['IGH','IGK','IGL'],
                 linewidth=3)

handles, labels = g.get_legend_handles_labels()
plt.legend(handles=handles[1:], labels=labels[1:], frameon=False, loc = 'lower center', ncol=3,
           bbox_to_anchor=(.5,-.3))
g.set(xlabel='Number of individuals sharing the clonotype sequence', ylabel='Number of clonotype sequences',
      yscale='log')
plt.xticks(range(1,16, 2), labels=['1','3','5','7','9','11','13','15+'])
plt.yticks([1,10,100,1000,10000,100000,1000000])

sns.despine()

plt.savefig('../figures/Figure4_a.png', bbox_inches='tight')
plt.savefig('../figures/Figure4_a.pdf', bbox_inches='tight')
```
