```python
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# read length 50 - sensitivity
fmri = pd.read_csv("sensitivity.50.csv", header=0)
fmri = pd.melt(fmri, id_vars=['tool'], value_vars=['1', '2', '4', '8', '16', '32', '64', '128'])
fmri.columns = ["Tool", "Coverage", "Sensitivity"]
fmri.astype({'Coverage': 'int32'}, copy = False).dtypes
plt.figure(figsize=(10, 10))
sns.set_style("whitegrid")
sns.set_context("poster", font_scale=1.5, rc={"lines.linewidth": 4})
ax = sns.lineplot(x="Coverage", y="Sensitivity", hue = "Tool", style="Tool", markers = True, 
                  data=fmri, sort = False)
ax.set_xlim(-0.2, 7.2)
z = ax.set_ylim(-0.05, 1.05)
title = plt.title('Sensitivity (read length 50)')
```
