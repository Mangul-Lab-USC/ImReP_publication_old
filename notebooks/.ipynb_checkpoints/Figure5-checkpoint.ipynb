```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats
from matplotlib.ticker import ScalarFormatter
```

```python
colors = ['#F7941D', '#00b9f2', '#00a875', '#ecde38', '#0072bc', '#F15a22', '#da6fab']
sns.set_palette(sns.color_palette(colors))
sns.set_context('talk')
sns.set_style('white')
```

```python
tissue_df = pd.read_csv('../summary_data/Figure5a.csv')
tissue_df = pd.melt(tissue_df, id_vars=['Chain type'], value_vars=['The same individuals', 'Different individuals'], 
                    var_name='sharing_type', value_name='average_shared')
tissue_df
```

```python
plt.figure(figsize=(7,5))
fig, ax = plt.subplots(nrows=1, ncols=1)

g=sns.barplot(data=tissue_df, x='Chain type', y='average_shared', hue='sharing_type', saturation=1)
g.set(xlabel='', ylabel='Number of shared clonotypes')

handles, labels = g.get_legend_handles_labels()
plt.legend(handles, ['across tissues within the same individual', 'across tissues from different individuals'], 
           bbox_to_anchor=(0.47, -.4, 0, 0.), frameon=False, ncol=1, handletextpad=0.2, columnspacing=0.8, 
           loc='lower center', prop={'size': 15})
g=sns.despine()

ax.yaxis.set_major_formatter(ScalarFormatter())
plt.yticks(range(0,15,2))

plt.savefig('../figures/Figure5_a.png', bbox_inches='tight')
plt.savefig('../figures/Figure5_a.pdf', bbox_inches='tight')
```

```python
df_raw=pd.read_csv('../summary_data/cdr3_perSample.csv')
df_raw.head()
```

```python
df_meta=pd.read_csv('../summary_data/perSample_allChains_aditionalPhenotypes.csv')
df_meta.head()
```

```python
df=pd.merge(df_raw,df_meta,on='File_Name')
df.head()
```

```python
tmp = df.drop_duplicates(['tissue_y', 'submitted_subject_id_s'])
tmp = tmp.groupby(['tissue_y']).count()['submitted_subject_id_s'].reset_index().rename(
    columns={'submitted_subject_id_s': 'num_subjects'})
toi = tmp.loc[tmp['num_subjects'] >= 10, 'tissue_y'].tolist()
df = df[df.tissue_y.isin(toi)]
```

```python
df_ligth=df[['File_Name','submitted_subject_id_s','tissue_x','histological site','type','cdr3']]
df_ligth['cdr3_ind']=df_ligth['submitted_subject_id_s'].astype(str)+'_'+df_ligth['cdr3']
df_ligth.head()
```

```python
df_IGK=df_ligth.loc[df_ligth['type'] == 'IGK']
df_IGK.head()
```

```python
df_IGK=df_IGK[['submitted_subject_id_s','cdr3']]
#remove dublicates, dublicates are due to CDR3 shared across tissues 
df_IGK_no_dublicates = df_IGK.drop_duplicates()
print ("No duplicates",df_IGK.shape)
print ("Duplicates",df_IGK_no_dublicates.shape)
```

```python
df_IGK_no_dublicates['n_across_ind'] =  df_IGK_no_dublicates.groupby('cdr3')['cdr3'].transform('size')
df_IGK_no_dublicates.head()
```

```python
df_IGK['n_across_samples'] =  df_IGK.groupby('cdr3')['cdr3'].transform('size')
df_IGK.head()
```

```python
df_IGK_final=pd.merge(df_IGK,df_IGK_no_dublicates,on='cdr3')
df_IGK_final.head()
```

```python
df_IGK_final=df_IGK_final[['cdr3','n_across_samples','n_across_ind']]
df_IGK_final=df_IGK_final.drop_duplicates()
df_IGK_final.head()
```

```python
df_IGK_final['n_across_tissues']=df_IGK_final['n_across_samples']-df_IGK_final['n_across_ind']
df_IGK_final['Ig'] = 'IGK'
df_IGK_final.head()
```

```python
df_IGH=df_ligth.loc[df_ligth['type'] == 'IGH']
df_IGH.head()
```

```python
df_IGH=df_IGH[['submitted_subject_id_s','cdr3']]
#remove dublicates, dublicates are due to CDR3 shared across tissues 
df_IGH_no_dublicates = df_IGH.drop_duplicates()
print ("No dublicates",df_IGH.shape)
print ("Dublicates",df_IGH_no_dublicates.shape)
```

```python
df_IGH_no_dublicates['n_across_ind'] = df_IGH_no_dublicates.groupby('cdr3')['cdr3'].transform('size')
df_IGH_no_dublicates.head()
```

```python
df_IGH['n_across_samples'] =  df_IGH.groupby('cdr3')['cdr3'].transform('size')
df_IGH.head()
```

```python
df_IGH_final=pd.merge(df_IGH,df_IGH_no_dublicates,on='cdr3')
df_IGH_final.head()
```

```python
df_IGH_final=df_IGH_final[['cdr3','n_across_samples','n_across_ind']]
df_IGH_final=df_IGH_final.drop_duplicates()
df_IGH_final.head()
```

```python
df_IGH_final['n_across_tissues']=df_IGH_final['n_across_samples']-df_IGH_final['n_across_ind']
df_IGH_final['Ig'] = 'IGH'
df_IGH_final.head()
```

```python
df_IGL=df_ligth.loc[df_ligth['type'] == 'IGL']
df_IGL.head()
```

```python
df_IGL=df_IGL[['submitted_subject_id_s','cdr3']]
#remove dublicates, dublicates are due to CDR3 shared across tissues 
df_IGL_no_dublicates = df_IGL.drop_duplicates()
print ("No duplicates",df_IGL.shape)
print ("Duplicates",df_IGL_no_dublicates.shape)
```

```python
df_IGL_no_dublicates['n_across_ind'] =  df_IGL_no_dublicates.groupby('cdr3')['cdr3'].transform('size')
df_IGL_no_dublicates.head()
```

```python
df_IGL['n_across_samples'] =  df_IGL.groupby('cdr3')['cdr3'].transform('size')
df_IGL.head()
```

```python
df_IGL_final=pd.merge(df_IGL,df_IGL_no_dublicates,on='cdr3')
df_IGL_final.head()
```

```python
df_IGL_final=df_IGL_final[['cdr3','n_across_samples','n_across_ind']]
df_IGL_final=df_IGL_final.drop_duplicates()
df_IGL_final.head()
```

```python
df_IGL_final['n_across_tissues']=df_IGL_final['n_across_samples']-df_IGL_final['n_across_ind']
df_IGL_final['Ig'] = 'IGL'
df_IGL_final.head()
```

```python
df_final = pd.concat([df_IGH_final, df_IGK_final, df_IGL_final])
df_final.head()
```

```python
cdr3_df = pd.read_csv('../summary_data/cdr3_perSample_clean.csv')
cdr3_df.head()
```

```python
meta_df = pd.read_csv('../summary_data/perSample_allChains_aditionalPhenotypes.csv')
meta_df = meta_df.loc[:, ['File_Name', 'submitted_subject_id_s']]
meta_df = meta_df.rename(columns={'File_Name': 'sample', 'submitted_subject_id_s': 'subject'})
meta_df.head()
```

```python
full_df = pd.merge(cdr3_df, meta_df, on = 'sample')
full_df.head()
```

```python
df_igh = full_df.loc[full_df['type'] == 'IGH', :]
df_igh.head()
```

```python
df_igk = full_df.loc[full_df['type'] == 'IGK', :]
df_igk.head()
```

```python
df_igl = full_df.loc[full_df['type'] == 'IGL', :]
df_igl.head()
```

```python
igh_samples = df_igh['sample'].drop_duplicates().tolist()
igh_samples[0:5]
```

```python
igk_samples = df_igk['sample'].drop_duplicates().tolist()
igk_samples[0:5]
```

```python
igl_samples = df_igl['sample'].drop_duplicates().tolist()
igl_samples[0:5]
```

```python
igh_out = pd.merge(df_igh, df_igh, on = 'cdr3')
igh_out = igh_out.loc[igh_out['site_x'] != igh_out['site_y'], :]
igh_out['check_string'] = igh_out.apply(lambda row: ''.join(sorted([row['sample_x'], row['sample_y']])), axis=1)
igh_out['tissue_pair'] = igh_out.apply(lambda row: ','.join(sorted([row['site_x'], row['site_y']])), axis=1)
igh_out = igh_out.drop_duplicates(['check_string', 'cdr3'])
igh_out.head()
```

```python
igh_same = igh_out.loc[igh_out['subject_x'] == igh_out['subject_y'], :]
igh_same_summ = igh_same.groupby('tissue_pair').count()['check_string'].reset_index()
igh_same_summ = igh_same_summ.rename(columns={'check_string': 'igh_same_ind'})
igh_same_summ.head()
```

```python
igh_diff = igh_out.loc[igh_out['subject_x'] != igh_out['subject_y'], :]
igh_diff_summ = igh_diff.groupby('tissue_pair').count()['check_string'].reset_index()
igh_diff_summ = igh_diff_summ.rename(columns={'check_string': 'igh_diff_ind'})
igh_diff_summ.head()
```

```python
print(len(df_igk))
print(len(df_igh))
```

```python
igk_out = pd.merge(df_igk, df_igk, on = 'cdr3')
igk_out = igk_out.loc[igk_out['site_x'] != igk_out['site_y'], :]
igk_out['check_string'] = igk_out.apply(lambda row: ''.join(sorted([row['sample_x'], row['sample_y']])), axis=1)
igk_out['tissue_pair'] = igk_out.apply(lambda row: ','.join(sorted([row['site_x'], row['site_y']])), axis=1)
igk_out = igk_out.drop_duplicates(['check_string', 'cdr3'])
igk_out.head()
```

```python
igk_same = igk_out.loc[igk_out['subject_x'] == igk_out['subject_y'], :]
igk_same_summ = igk_same.groupby('tissue_pair').count()['check_string'].reset_index()
igk_same_summ = igk_same_summ.rename(columns={'check_string': 'igk_same_ind'})
igk_same_summ.head()
```

```python
igk_diff = igk_out.loc[igk_out['subject_x'] != igk_out['subject_y'], :]
igk_diff_summ = igk_diff.groupby('tissue_pair').count()['check_string'].reset_index()
igk_diff_summ = igk_diff_summ.rename(columns={'check_string': 'igk_diff_ind'})
igk_diff_summ.head()
```

```python
igl_out = pd.merge(df_igl, df_igl, on = 'cdr3')
igl_out = igl_out.loc[igl_out['site_x'] != igl_out['site_y'], :]
igl_out['check_string'] = igl_out.apply(lambda row: ''.join(sorted([row['sample_x'], row['sample_y']])), axis=1)
igl_out['tissue_pair'] = igl_out.apply(lambda row: ','.join(sorted([row['site_x'], row['site_y']])), axis=1)
igl_out = igl_out.drop_duplicates(['check_string', 'cdr3'])
igl_out.head()
```

```python
igl_same = igl_out.loc[igl_out['subject_x'] == igl_out['subject_y'], :]
igl_same_summ = igl_same.groupby('tissue_pair').count()['check_string'].reset_index()
igl_same_summ = igl_same_summ.rename(columns={'check_string': 'igl_same_ind'})
igl_same_summ.head()
```

```python
igl_diff = igl_out.loc[igl_out['subject_x'] != igl_out['subject_y'], :]
igl_diff_summ = igl_diff.groupby('tissue_pair').count()['check_string'].reset_index()
igl_diff_summ = igl_diff_summ.rename(columns={'check_string': 'igl_diff_ind'})
igl_diff_summ.head()
```

```python
df_final = pd.merge(igh_same, igh_diff, on = 'tissue_pair', how='outer')
df_final = pd.merge(df_final, igk_same, on = 'tissue_pair', how='left')
df_final = pd.merge(df_final, igk_diff, on = 'tissue_pair', how='left')
df_final = pd.merge(df_final, igl_same, on = 'tissue_pair', how='left')
df_final = pd.merge(df_final, igl_diff, on = 'tissue_pair', how='left')

df_final.to_csv(path_or_buf='../summary_data/Figure5a_data.csv')
```
