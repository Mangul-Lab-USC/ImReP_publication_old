```python
import pandas as pd
```

```python
df=pd.read_csv('../summary_data/data_hashimotos.csv')
df=df.loc[df['type'] != 'focal']
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

g=sns.set_style("white")
g=sns.set_context("poster")
g=sns.catplot(x="type", y="n_cdr3_IGH",data = df,kind='bar',height=5,aspect=3.5,order=['None','patchy','early',' hashimotos'])

g.set(xlabel='Severity of Hashimoto\'s thyroiditis', ylabel='Number of IGH clonotypes')
#plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
#plt.ylim(-1, 1)
g=sns.despine()
plt.savefig("Severity_hashimoto.pdf", bbox_inches='tight')
```
